```python
# 1. Write a program that takes a string input from the user and prints the length of the string.
s = input("Enter a string: ")
print("Has escrito:", s)
print("Longitud:", len(s))
```

```python
# 2. Create a program that converts a given string to uppercase.
# s = input("Enter a string: ")
s_mayus = s.upper()
print("Original:", s)
print("Mayúsculas:", s_mayus)
```

```python
# 3. Write a function that checks if a string contains only alphabets.
# s = input("Enter a string: ")
solo_letras = True
for c in s:
    if c.isalpha() == False:
        solo_letras = False
        break  # ya no hace falta seguir mirando
print("Solo letras:", solo_letras)
```

```python
# 4. Reverse a string without using slicing (e.g., string[::-1]).
# s = input("Enter a string: ")
reversed_s = ""
for c in s:
    reversed_s = c + reversed_s
print("Reversed string:", reversed_s)
```

```python
# 5. Count how many times a specific character appears in a string.
# s = input("Enter a string: ")
characters_to_count = input("Enter the character you want to count:")
count = 0
for c in s:
    if c == characters_to_count:
        count += 1
print("The character", characters_to_count, "apeares", count, "times.")
```

```python
# 6. Write a function that takes two strings and checks if they are anagrams.
s1 = input("Enter a string: ")
print(s1)
s2 = input("Enter another string: ")
print(s2)
if sorted(s1) == sorted(s2):
    print("The strings are anagrams.")
else:
    print("The strings are not anagrams.")
```

```python
# 7. Implement a function that removes all vowels from a given string.
vowels = "aeiouAEIOU"
no_vowels = ""
for c in s:
    if c not in vowels:
        no_vowels += c
print("Original string:", s)
print("String without vowels:", no_vowels)
```

```python
# 8. Write a function that checks if a string is a palindrome (case-insensitive).
s = s.lower()
izq = 0
der = len(s) - 1

es_palindromo = True
while izq < der:
    if s[izq] != s[der]:
        es_palindromo = False
        break
    izq = izq + 1
    der = der - 1
print("Es palíndromo:", es_palindromo)
```

```python
# 9. Given a string, return the first non-repeating character.
s = input("Enter a string: ")
characters_count = {}
for c in s:
    if c in characters_count:
        characters_count[c] += 1
    else:
        characters_count[c] = 1
        not_repeated = None
for c in s:
    if characters_count[c] == 1:
        not_repeated = c
        break

print("First character that isn't repeated:", not_repeated)
```

```python
# 10. Create a function that capitalizes the first letter of each word in a string.
s = input("Enter a string:")
words = s.split(" ")
capitalized_words =[]

for w in words:
    capitalized_words.append(w.capitalize())
    capitalized_s = " ".join(capitalized_words)

print("Capitalized string:", capitalized_s)
```

```python
# 2. Write a program to merge two dictionaries.
dict1 = {"a": 1, "b": 2, "d": 3}
dict2 = {"c":4, "e":5, "f":7}

dict3 = {}
for k in dict1:
    dict3[k] = dict1[k]
for k in dict2:
    dict3[k] = dict2[k]
    
print("Merged dictionary:", dict3)
```

```python
# 3. Implement a function to count the occurrences of each word in a given sentence.
sentence = input("Enter a sentence: ")
words = sentence.split(" ")
word_count = {}

for w in words:
    if w in word_count:
        word_count[w] += 1
    else:
        word_count[w] = 1

print("Word occurrences:", word_count)
```
