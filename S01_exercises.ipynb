```python
# 8. Write a function that checks if a string is a palindrome (case-insensitive).
s = s.lower()
izq = 0
der = len(s) - 1

es_palindromo = True
while izq < der:
    if s[izq] != s[der]:
        es_palindromo = False
        break
    izq = izq + 1
    der = der - 1
print("Es palíndromo:", es_palindromo)
```

```python
# 9. Given a string, return the first non-repeating character.
s = input("Enter a string: ")
characters_count = {}
for c in s:
    if c in characters_count:
        characters_count[c] += 1
    else:
        characters_count[c] = 1
        not_repeated = None
for c in s:
    if characters_count[c] == 1:
        not_repeated = c
        break

print("First character that isn't repeated:", not_repeated)
```

```python
# 10. Create a function that capitalizes the first letter of each word in a string.
s = input("Enter a string:")
words = s.split(" ")
capitalized_words =[]

for w in words:
    capitalized_words.append(w.capitalize())
    capitalized_s = " ".join(capitalized_words)

print("Capitalized string:", capitalized_s)
```

```python
# 11. Write a program to compress a string (e.g., "aaabb" -> "a3b2").
s = input("Enter a string:")
compressed_s = ""
i = 0
while i < len(s):
    count = 1
    while i + 1 < len(s) and s[i] == s[i + 1]:
        count += 1
        i += 1
    compressed_s += s[i] + str(count)
    i += 1
    print("Compressed string:", compressed_s)
```

```python
# 12. Implement a function to find the longest substring without repeating characters.
def longest_substring(s):
    start = ""                      # start of the longest substring found
    n = len(s)                      # lenght of the string

    for i in range(n):              # i = position where the substring starts
        seen = set()                # letters used in the current substring

        for j in range(i, n):                   # j = position where the substring ends
            letter = s[j]                       # letter in j position
            
            if letter in seen:                  # if the letter is already used, repeat
                break                           # stop the inner loop

            seen.add(letter)                    # if not, add the letter to the set
            candidate = s[i:j+1]                #  current substring
            if len(candidate) > len(start):
                start = candidate               # update the longest substring found
    return start

s = input("Enter a string:")
longest = longest_substring(s)
print("Longest substring without repeating characters:", longest)
```

```python
# 1. Create a dictionary to store the squares of numbers from 1 to 10.
cuadráticas = {}
for i in range(1,11):
    cuadráticas[i] = i*i
print(cuadráticas)
```

```python
# 2. Write a program to merge two dictionaries.
dict1 = {"a": 1, "b": 2, "d": 3}
dict2 = {"c":4, "e":5, "f":7}

dict3 = {}
for k in dict1:
    dict3[k] = dict1[k]
for k in dict2:
    dict3[k] = dict2[k]
    
print("Merged dictionary:", dict3)
```

```python
# 3. Implement a function to count the occurrences of each word in a given sentence.
sentence = input("Enter a sentence: ")
words = sentence.split(" ")
word_count = {}

for w in words:
    if w in word_count:
        word_count[w] += 1
    else:
        word_count[w] = 1

print("Word occurrences:", word_count)
```

```python
# 4. Create a dictionary from two lists: one for keys and one for values.
keys = ["a", "b", "c", "d"]
values = [1, 2, 3, 4]

dict_keys_values = {}
longitud = len(keys)
for i in range(longitud):
    dict_keys_values[keys[i]] = values[i]

print("Dictionary from keys and values:", dict_keys_values)
```

```python
# 5. Write a program to check if a given key exists in a dictionary.
key_check = input("Enter a key to check:")
if key_check in dict_keys_values:
    print("The key", key_check, "exists in the current dictionary")
else:
    print("The key", key_check, "does not exist in the current dictionary")
```

```python
# 6. Implement a function to invert a dictionary (swap keys and values).
d = {"a": 1, "b": 2, "c": 3}
def invert_dict(d):
    inverted = {}
    for key, value in d.items():
        inverted[value] = key # swap keys and values
    return inverted

inverted_d = invert_dict(d)
print("Inverted dictionary:", inverted_d)
```

```python
# 7. Write a program to find the key with the highest value in a dictionary.
d = {"a": 1, "b": 2, "c": 3, "d": 4, "e": 5}
def key_highest_value(d):
    max_key = None
    max_value = None

    for key, value in d.items():
        if max_value is None or value > max_value:
            max_value = value
            max_key = key
        
    return max_key

key_max = key_highest_value(d)
print("Key with highest value:", key_max)
```
